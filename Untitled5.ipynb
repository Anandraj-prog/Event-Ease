```python
import numpy as np
import torch
import matplotlib.pyplot as plt
import cv2
import os
from segment_anything import sam_model_registry, SamAutomaticMaskGenerator, SamPredictor

# Define paths
sam_checkpoint = "sam_vit_h_4b8939.pth"  # Path to the SAM checkpoint (download from Meta's repository)
image_path = "dog.jpg"     # Path to your input image

# Function to process large images by tiling them into chunks (for GPU VRAM limitations)
def process_large_image(image, tile_size=1024, overlap=128):
    h, w, _ = image.shape
    tiles = []

    for y in range(0, h, tile_size - overlap):
        for x in range(0, w, tile_size - overlap):
            y_end = min(y + tile_size, h)
            x_end = min(x + tile_size, w)
            tile = image[y:y_end, x:x_end]
            tiles.append((tile, (x, y)))

    return tiles

# Load the image with error checking
image = cv2.imread(image_path)
if image is None:
    raise FileNotFoundError(f"Image file not found at {image_path}. Please check the file path.")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Load SAM model
model_type = "vit_h"
device = "cuda" if torch.cuda.is_available() else "cpu"

sam = sam_model_registry[model_type](checkpoint=sam_checkpoint)
sam.to(device=device)

# Generate masks for all prominent objects
mask_generator = SamAutomaticMaskGenerator(sam)

# Handle large images by processing them in tiles if the image size is large
image_height, image_width, _ = image.shape
max_image_size = 2048  # Max size for a single image tile (for GPU memory management)
if image_height > max_image_size or image_width > max_image_size:
    print("Large image detected, processing in tiles...")
    tiles = process_large_image(image)
else:
    tiles = [(image, (0, 0))]

# Output directory for segmented objects
output_dir = "segmented_objects"
os.makedirs(output_dir, exist_ok=True)

# Process each tile and generate masks
masks_all = []
for tile, (x_offset, y_offset) in tiles:
    masks = mask_generator.generate(tile)
    masks_all.extend(masks)

    # Output each segmented object as a separate image file with a transparent background
    for i, mask in enumerate(masks):
        segmentation = mask['segmentation']

        # Create an alpha channel (255 for masked regions and 0 for others)
        alpha_channel = np.zeros(segmentation.shape, dtype=np.uint8)
        alpha_channel[segmentation] = 255  # Fully opaque where the mask is applied

        # Create RGBA image (original image + alpha channel)
        rgba_image = np.dstack((tile, alpha_channel))  # Combine RGB with alpha channel

        # Output path with adjusted naming convention for tiles
        output_path = os.path.join(output_dir, f'segmented_object_{x_offset}_{y_offset}_{i}.png')
        cv2.imwrite(output_path, cv2.cvtColor(rgba_image, cv2.COLOR_RGB2BGRA))  # Convert to BGRA format for transparency

# Function to display masks (with transparency)
def show_anns(anns):
    if len(anns) == 0:
        return
    sorted_anns = sorted(anns, key=(lambda x: x['area']), reverse=True)
    ax = plt.gca()
    ax.set_autoscale_on(False)
    img = np.ones((sorted_anns[0]['segmentation'].shape[0], sorted_anns[0]['segmentation'].shape[1], 4))
    img[:,:,3] = 0  # Set alpha channel to 0 (transparent)
    for ann in sorted_anns:
        m = ann['segmentation']
        color_mask = np.concatenate([np.random.random(3), [0.35]])  # Add random color with transparency
        img[m] = color_mask
    ax.imshow(img)

# Display the masks on the original image
plt.figure(figsize=(20,20))
plt.imshow(image)
show_anns(masks_all)
plt.axis('off')
plt.show()

# Handle large images by tiling or chunking (if needed for performance)
mask_generator_2 = SamAutomaticMaskGenerator(
    model=sam,
    points_per_side=32,
    pred_iou_thresh=0.86,
    stability_score_thresh=0.92,
    crop_n_layers=1,
    crop_n_points_downscale_factor=2,
    min_mask_region_area=100,  # Requires open-cv to run post-processing
)

masks2 = mask_generator_2.generate(image)

# Display second round of masks
plt.figure(figsize=(20,20))
plt.imshow(image)
show_anns(masks2)
plt.axis('off')
plt.show()



```
